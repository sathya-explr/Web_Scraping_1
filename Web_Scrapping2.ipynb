```python
#convert the web page to text
Link_text = requests.get(link).text
```

```python
Link_text
```

```python
#to convert Link_text into a BeautifulSoup Object
soup1 = BeautifulSoup(Link_text, 'lxml')
print(soup1)
```

```python
#make the indentation proper
print(soup1.prettify())
```

```python
#Fetch all the table tags
all_table = soup.find_all('table')
print(all_table)
```

```python
# here we extract the table using table class name
# 
our_table = soup1.find('table', class_= 'table table-striped table-hover ht-table table-responsive-md table-bordered')
```

```python
our_table
```

```python
# it works with the first bs object also
our_table = soup.find('table', class_= 'table table-striped table-hover ht-table table-responsive-md table-bordered')
```

```python
our_table
```

```python
type(our_table)
```

```python
# We try to get rows of the table
allrows = our_table.find_all('tr')
```

```python
allrows
```

```python
# now we have table in the form of list of rows
# so we can get data and header seprately
data =[]

# get data in each row using for loop
for row in allrows:
    row_list = row.find_all('td')
    # have a list for data in each row
    row_data = []
    # adding the cell data to the list
    for cell in row_list:
       row_data.append(cell.text)
# now every row's data in a separate list, we create a list of lists
    data.append(row_data)
print(data)
```

```python
data[1]
# we see that here actual table rows start
```

```python
#We can use this list to create a dataframe
df = pd.DataFrame(data[1:])
```
