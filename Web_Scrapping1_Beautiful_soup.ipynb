```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
import requests
```

```python
link = "https://www.hubertiming.com/results/2022Stumptown4"
```

```python
page_read = requests.get(link)
```

```python
page_read
```

```python
soup = BeautifulSoup(page_read.content)
```

```python
soup
```

```python
# get title
title = soup.title
```

```python
title
```

```python
# to get only text of title without html

title.text

# we can also use print(soup.title.string)
```

```python
# to get links
links = soup.find_all('a' , href=True)
```

```python
print(links)
```

```python
# to get the exact link paths alone
links = soup.find_all('a', href=True)
for link in links:
    print(link['href'])
```

```python
# using the get method
for link in links:
    print(link.get('href'))
```

```python
# finding tr and td tags
allrows = soup.find_all('tr')
# allrows is a list
for row in allrows:
    row_list = row.find_all('td')
# print the last row data   
print(row_list)
```

```python
# to see only text part of the row without html tags
for data in row_list:
    print(data.text)
```

```python
# Getting the table data when there is only one table on the webpage
data = []
allrows = soup.find_all('tr')
# get data in each row using for loop
for row in allrows:
    row_list = row.find_all('td')
    # have a list for data in each row
    row_data = []
    # adding the cell data to the list
    for cell in row_list:
       row_data.append(cell.text)
# now every row's data in a separate list, we create a list of lists
    data.append(row_data)
print(data)
```

```python
# the list data contains all the table rows as a list
data[1]
# gives the element at first position
```

```python
data[5]
# from the webpage we see that this is where the table rows start.
```

```python
# so from index position 5, we convert the list to dataframe
df = pd.DataFrame(data[5:])
```

```python
df.head()
```

```python
# We find the column headers using the th tag

col_headers = soup.find_all('th')
col_headers

# these are the column names    
```

```python
# use a list to get the text value of col_headers 
header_list = []
for col in col_headers:
    header_list.append(col.text)
header_list
```

```python
# now we can directly assign it to our df column
df.columns = header_list
```

```python
df.head()
```
